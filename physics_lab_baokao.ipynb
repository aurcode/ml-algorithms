```
import numpy as np
import pandas as pd

# Given data
h_values = np.array([10, 20, 30, 40, 50, 60])  # in mm
trial_1 = np.array([14.000, 24, 25, 30, 31, 36])  # in ms
trial_2 = np.array([19.000, 22, 25, 27, 30, 39])  # in ms
trial_3 = np.array([13.000, 21, 28, 26, 29, 33])  # in ms

# Compute average T values (in ms)
T_avg = (trial_1 + trial_2 + trial_3) / 3

# Calculate T^2 (in ms^2)
T_squared = T_avg ** 2

# Fit T^2 as a linear function of h to find slope (k)
slope, intercept = np.polyfit(h_values, T_squared, 1)

# Calculate  (adiabatic index) using k = (4^2d^2)/( * P * m)
# Constants
d = 32.5e-3  # diameter in meters
P = 101300  # pressure in Pascals
m = 0.0405  # mass in kg

gamma = (4 * np.pi**2 * d**2) / (slope * P * m)

# Create a DataFrame to organize the results
results = pd.DataFrame({
    "h (mm)": h_values,
    "T_avg (ms)": T_avg,
    "T^2 (ms^2)": T_squared
})

results.loc["Slope (k)", :] = ["-", "-", slope]
results.loc["", :] = ["-", "-", gamma]

results
```
